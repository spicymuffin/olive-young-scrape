## 감성분석

```python
import re

with open('reviews.txt', 'r', encoding='utf-8') as f:
    reviews = f.readlines()

positive_keywords = ['좋아요', '예뻐요', '추천', '만족', '촉촉']
negative_keywords = ['싫어요', '아쉬워요', '불만족', '건조', '실망']

labeled_reviews = []
for review in reviews:
    review_cleaned = re.sub(r'<[^>]*>', '', review.strip())  
    label = 0  
    if any(word in review_cleaned for word in positive_keywords):
        label = 1
    elif any(word in review_cleaned for word in negative_keywords):
        label = 0
    labeled_reviews.append(f"{review_cleaned}\t{label}")

with open('labeled_reviews.txt', 'w', encoding='utf-8') as f:
    f.write('\n'.join(labeled_reviews))

print("labeled_reviews.txt")
```

## Topic modeling

```python
def get_highest_topic(topic_list):
    highest_topic = 100
    highest_prob = 0
    for topic, prob in topic_list:
        if prob > highest_prob:
            highest_prob = prob
            highest_topic = topic
    return highest_topic, highest_prob
```

```python
def get_adjective_words(morphs):
    adjective_words = []
    for word, tag in morphs:
        if tag == 'Adjective':  
            adjective_words.append(word)
    return adjective_words
```

```python
from kiwipiepy import Kiwi

# Kiwi 초기화
kiwi = Kiwi()

# 형용사 추출 함수 정의
def get_adjective_words(text):
    # 텍스트를 분석하고 첫 번째 결과에서 토큰 리스트 추출
    tokens = kiwi.analyze(text)[0][0]  
    # 품사가 'VA'(형용사)인 토큰만 필터링하여 단어 리스트 생성
    adjectives = [token[0] for token in tokens if token[1] == 'VA']  
    return adjectives

# 결과 저장 리스트 준비
documents = []  # 형용사를 저장할 리스트
article_ids = []  # 문서 ID를 저장할 리스트
text_titles = []  # 문서 제목을 저장할 리스트

# total_morphs를 순회하면서 형용사 추출
for key in total_morphs:
    # 각 데이터의 'cleaned_text' 필드에서 텍스트 가져오기
    cleaned_text = total_morphs[key]['cleaned_text']
    # 형용사를 추출하고 documents 리스트에 추가
    documents.append(get_adjective_words(cleaned_text))

    # paper_id와 text_title 필드가 존재하지 않을 경우 'N/A'를 기본값으로 설정
    article_ids.append(total_morphs[key].get('paper_id', 'N/A'))
    text_titles.append(total_morphs[key].get('text_title', 'N/A'))

# 결과 출력 (처음 5개만 출력)
print("추출된 형용사 (처음 5개):", documents[:5])
print("문서 ID (처음 5개):", article_ids[:5])
print("문서 제목 (처음 5개):", text_titles[:5])
```

```python
# 불용어 제거 => 불용어 사전 + 1음절 단어
docs_filtered = [[term for term in doc if term not in stop_words and len(term)>1] for doc in documents]
```

```python
print(docs_filtered[0])
```

```python
from gensim.corpora import Dictionary
dictionary = Dictionary(docs_filtered)
```

```python
dictionary.filter_extremes(no_below=10, no_above=0.1)
```

```python
DTM = []
for doc in docs_filtered:
    bow = dictionary.doc2bow(doc)
    DTM.append(bow)
```

```python
print(dictionary.token2id)
```

```python
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(DTM))
```

```python
print(DTM[0])
```

```python
from gensim import models
```

```python
NUM_TOPICS = 30
```

```python
lda_model = models.ldamodel.LdaModel(corpus=DTM, num_topics=NUM_TOPICS,
                        id2word=dictionary, alpha='auto', eta='auto', iterations = 1000)
```

```python
TOPIC_ID = 0
NUM_TOPIC_WORDS = 10
lda_model.show_topic(TOPIC_ID, NUM_TOPIC_WORDS)
```

```python
for topic_id in range(lda_model.num_topics):
    word_probs = lda_model.show_topic(topic_id, NUM_TOPIC_WORDS)
    print("Topic ID: {}".format(topic_id))
    for word, prob in word_probs:
        print("\t{}\t{}".format(word, prob))
    print("\n")
```

```python
# get_highest_topic 함수 정의
def get_highest_topic(topics_list):
    # topics_list에서 확률이 가장 높은 주제를 찾음
    hi_topic, hi_prob = max(topics_list, key=lambda x: x[1])  # 확률 기준으로 정렬하여 최대값 선택
    return hi_topic, hi_prob

# 10개의 문서를 순회하며 주제 출력
for i in range(10):
    print(article_ids[i])  # 문서 ID 출력
    print(text_titles[i][:50] + "...")  # 문서 제목을 50자까지만 출력하고 나머지는 생략
    
    # 문서의 주제 분포 가져오기
    topics_list = lda_model.get_document_topics(DTM[i], minimum_probability=0.05)
    print(topics_list)  # 주제 분포 출력

    # topics_list가 비어 있는 경우 처리
    if not topics_list:
        print("No topics found for this document.\n")
        continue  # 다음 문서로 넘어감

    # 확률이 가장 높은 주제 찾기
    hi_topic, hi_prob = get_highest_topic(topics_list)
    print(f"The most covered topic is Topic {hi_topic} and percentage is {hi_prob:.2f}\n")
```

```python
start1 = 3
limit1 = 50
step1 = 5
perplexity_scores=[]
for num_topics in range(start1, limit1, step1):
    model = models.ldamodel.LdaModel(DTM, num_topics=num_topics,
                    id2word=dictionary, alpha='auto', eta='auto', iterations=100)
    perplexity_scores.append(model.log_perplexity(DTM))
```

```python
import matplotlib.pyplot as plt
```

```python
x = range(start1, limit1, step1)
plt.plot(x, perplexity_scores)
plt.xlabel("Num Topics")
plt.ylabel("Log Perplexity scores")
plt.legend(("Log Perplexity"), loc='best')
plt.show()
```

```python
from gensim.models import CoherenceModel
```

```python
def compute_coherence_values_umass(dictionary, corpus, texts, limit, start=2, step=3):
    coherence_values = []
    for num_topics in range(start, limit, step):
        model = models.ldamodel.LdaModel(corpus, num_topics=num_topics,
                        id2word=dictionary, alpha='auto', eta='auto')
        coherencemodel = CoherenceModel(model=model, texts=texts, dictionary=dictionary, 
                                        coherence='u_mass', topn=30)
        coherence_values.append(coherencemodel.get_coherence())

    return coherence_values
```

```python
start1 = 3
limit1 = 63
step1 = 3
coherence_values_umass = compute_coherence_values_umass(dictionary=dictionary, corpus=DTM, texts=docs_filtered, start=start1, limit=limit1, step=step1)
```

```python
x = range(start1, limit1, step1)
plt.plot(x, coherence_values_umass)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score: u_mass")
plt.legend(("coherence_values"), loc='best')
plt.show()
```
